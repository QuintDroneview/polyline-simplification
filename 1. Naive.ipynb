```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
poly = np.array([[0,0],
                 [0.5,0.5],
                 [1,0],
                 [4,0],
                 [6,1],
                 [6.5,2],
                 [9,-1],
                 [10,4]
                ]);
#poly = poly0
fig = plt.figure(figsize=(15,6))
fig.gca().set_aspect('equal', adjustable='box')
plt.plot(poly[:,0],poly[:,1])
plt.scatter(poly[:,0],poly[:,1])
```

```python
def simplify_naive(n,p):
    mask = np.zeros(len(p),dtype='bool')
    mask[0:len(p):n] = True
    mask[-1] = True
    return mask
    
mask = simplify_naive(2,poly)
simplified = poly[mask]

#preparing the figure
plt.figure(figsize=(15,6))
fig.gca().set_aspect('equal', adjustable='box')

#draw the original polygon
plt.plot(poly[:,0],poly[:,1])
plt.scatter(poly[:,0],poly[:,1])

#draw the simplified polygon
plt.plot(simplified[:,0],simplified[:,1])
plt.scatter(simplified[:,0],simplified[:,1])

#scatter the deleted points
deleted = poly[np.logical_not(mask)]
plt.scatter(deleted[:,0],deleted[:,1],color='red',marker='x',s=50,linewidth=2.0)
```

```python
def plotpoly(n):
    mask = simplify_naive(n,poly)
    simplified = poly[mask]

    #preparing the figure
    plt.figure(figsize=(15,6))
    fig.gca().set_aspect('equal', adjustable='box')

    #draw the original polygon
    plt.plot(poly[:,0],poly[:,1])
    plt.scatter(poly[:,0],poly[:,1])

    #draw the simplified polygon
    plt.plot(simplified[:,0],simplified[:,1])
    plt.scatter(simplified[:,0],simplified[:,1])

    #scatter the deleted points
    deleted = poly[np.logical_not(mask)]
    plt.scatter(deleted[:,0],deleted[:,1],color='red',marker='x',s=50,linewidth=2.0)

from IPython.html.widgets import *
interact(plotpoly, n=(1,10,1));
```
